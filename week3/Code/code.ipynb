```python
import os
import pandas as pd
```

```python
folder_path  = "D:\Shair storae\Project\DE Zoom camp\DE\Module 3\Data"
```

```python
parquet_files = [f for f in os.listdir(folder_path) if f.endswith('.parquet')]
```

```python
parquet_files
```

```python
dfs = []
for file in parquet_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_parquet(file_path)
    dfs.append(df)
```

```python
combined_df = pd.concat(dfs, ignore_index=True)
```

```python
combined_df
```

```python
pathT = F"D:\Shair storae\Project\DE Zoom camp\DE\Module 3\compain\output.parquet" 
combined_df.to_parquet(pathT,index=False)
```
